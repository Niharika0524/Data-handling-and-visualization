```python
z=[20,30,40,50]
mylabels=["cat","dog","hare","dear"]
colours=["red","green","hotpink","pink"]
myexplode=[0,0,0.2,0.2]
plt.pie(z,labels=mylabels,colors=colours,explode=myexplode)
plt.show()
```

```python
import numpy as np
x=np.array(["A","B","C","D"])
y=np.array([3,8,1,10])
mycolours=["red","black","pink","hotpink"]
plt.bar(x,y,color=mycolours)
plt.show()
```

```python
x=np.array(["A","B","C","D"])
y=np.array([2,4,6,8])
mycolours=["red","black","pink","hotpink"]
sizes=[150,300,200,250]
plt.scatter(x,y,color=mycolours,s=sizes,alpha=0.9,marker="*",edgecolor="g")
plt.show()
```

```python
import matplotlib.pyplot as plt
x=[1,2,3,4]
y=[4,5,6,7]
plt.subplot(2,2,1)
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title("Line graph")
plt.plot(x,y)
plt.plot(x,y,color="r")
plt.subplot(2,2,2)
plt.pie([2],colors="r")
z=[20,30,40,50]
plt.subplot(2,2,3)
mylabels=["cat","dog","hare","dear"]
colours=["red","green","hotpink","pink"]
myexplode=[0,0,0.2,0.2]
plt.pie(z,labels=mylabels,colors=colours,explode=myexplode)
x=np.array(["A","B","C","D"])
y=np.array([3,8,1,10])
plt.subplot(2,2,4)
mycolours=["red","black","pink","hotpink"]
plt.bar(x,y,color=mycolours)
plt.show()


```
