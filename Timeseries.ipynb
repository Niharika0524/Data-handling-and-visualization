```python
import pandas as pd
import matplotlib.pyplot as plt
data={"Data":pd.to_datetime(["2022-01-01","2022-02-01","2022-03-01","2022-04-01","2023-01-01","2023-02-01","2023-03-01"]),"value":[15,22,28,33,42,48,55]}
df=pd.DataFrame(data)
df.set_index("Data",inplace=True)
plt.figure(figsize=(10,6))
plt.plot(df.index,df["value"],marker="*")
plt.xlabel("Date")
plt.ylabel("value")
plt.title("Time series with trend")
plt.grid(True)
plt.show()
```
