```python
import requests
link="https://oswalpublishers.com/blog/indian-scientists/"
link_text=requests.get(link).text
print(link_text)
```

```python
from bs4 import BeautifulSoup
soup=BeautifulSoup(link_text,'lxml')
print(soup)
```

```python
import requests
from bs4 import BeautifulSoup

def getdata(url):
    r = requests.get(url)
    return r.text

htmldata = getdata("https://oswalpublishers.com/blog/indian-scientists/")
soup = BeautifulSoup(htmldata, 'html.parser')
for item in soup.find_all('img'):
    print(item['src'])
```
